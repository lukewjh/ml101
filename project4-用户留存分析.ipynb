```python
# 导入数据 
import numpy as np
import pandas as pd
df_member = pd.read_csv('易速鲜花会员留存.csv')
df_member.head()
```

```python
# 在数据挖掘前一般要看看数据的样子
# 数据可视化开始
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号
plt.figure(figsize=(10,8))
# 分别创建子图
plt.subplot(2, 2, 1)
ax = df_member.groupby('性别').count()['用户码'].plot.pie(autopct='%1.0f%%')

plt.subplot(2, 2, 2)
ax = df_member.groupby('会费支付方式').count()['用户码'].plot.pie(autopct='%1.0f%%')

plt.subplot(2, 2, 3)
ax = df_member.groupby('会员卡类型').count()['用户码'].plot.pie(autopct='%1.0f%%')

plt.subplot(2, 2, 4)
ax = df_member.groupby('已停付会费').count()['用户码'].plot.pie(autopct='%1.0f%%')

plt.show()
```

```python
# 好了 可视化完了 就可以特征工程了
# 1.先将值为汉字的值转为数字
df_member['已停付会费'].replace(to_replace='是', value=1, inplace=True)
df_member['已停付会费'].replace(to_replace='否', value=0, inplace=True)
df_member['性别'].replace(to_replace='女', value=0, inplace=True)
df_member['性别'].replace(to_replace='男', value=1, inplace=True)

```

```python
# 套餐类型值得转成boolean值
binary_feature = ['玫瑰套餐', '紫罗兰套餐', '郁金香套餐', '百合套餐', '康乃馨套餐', 
                  '胡姬花套餐', '生日套餐','情人节套餐']
for field in binary_feature:
    df_member[field] = df_member[field] == '是'
```

```python
# 第二个可视化工作
# 显示会员得留存曲线 这个曲线在留存分析中非常重要 
# ******* Kaplan-Meier生存模型：显示整体留存曲线
# 在python中绘制留存曲线 可以使用一个很好用的
```

#### lifelines （生命线库）
- 优点在于简单、快捷、还能显示整个数据集中用户随着时间推移的留存情况
- 历史：lifelines最早是由精算师和医疗专业人员开发的，通过病人的病情和其他特征，来分析病人的生存可能性。后来，数据分析师们发现，这个模型既然能够判断哪些因子会影响病人的存活率（死亡率），那么，要是把因子换成用户行为特征和产品特征，不就也能分析出用户的留存率（流失率）吗？于是，这个工具被运营人员和数据分析师利用了起来，为分析用户留存提供了很多有用的方法。

```python
# 在lifelines工具包中有一个Kaplan-meier生存模型
# 能显示整体的用户留存曲线
# 重点：通过绘制不同群体的留存曲线我们就可以观察到数据集中的各个特征对于留存的影响大小

# 我们的数据中 最能够反应“留存时间“的字段有  已停付会费 、入会月数
# 这样就知道这个用户的生命周期长度了
```

```python
# 留存分析图
# 使用Kaplan-Meier生存模型 查看普通用户随着时间而变化的留存率
import lifelines
kmf = lifelines.KaplanMeierFitter()
kmf.fit(df_member['入会月数'], 
        event_observed=df_member['已停付会费'],
        label='会员预期留存线')
fig, ax = plt.subplots(figsize=(10,6))
kmf.plot(ax=ax)
#图题
ax.set_title('Kaplan-Meier留存曲线-易速鲜花会员们') 
#X轴标签
ax.set_xlabel('入会月数') 
#Y轴标签
ax.set_ylabel('留存率(%)') 
#显示图片
plt.show() 
```

```python
'''
如上，图中的曲线有蓝色阴影，这个阴影就是置信区间。
置信区间是一个范围，比如说有一批学生的考试成绩时从0-100
如果我们给出一个分数的区间[55分-95分]，然后说 “该区间的置信水平为95%”
这就代表学生的考试成绩落在这个范围的概率是95%
这里的[55-95]就被称做置信水平为95%的学生考试成绩的置信区间
'''
```

```python
# 根据不同的用户特征分组绘制留存曲线 创建一个绘制的函数
def life_by_cat(feature, t='入会月数', event='已停付会费', df=df_member, ax=None):
    for cat in df[feature].unique():
        idx = df[feature] == cat
        kmf = lifelines.KaplanMeierFitter()
        kmf.fit(df[idx][t], event_observed=df[idx][event], label=cat)
        kmf.plot(ax=ax, label=cat)
```

```python
# 先看看会员费缴纳方式对存留的影响
fig_pmt, ax_pmt = plt.subplots(figsize=(10,6))
life_by_cat(feature='会费支付方式', ax=ax_pmt)
ax_pmt.set_title('会费支付方式对留存的影响')
ax_pmt.set_xlabel('入会月数')
ax_pmt.set_ylabel('留存率（%）')
plt.show()
```

```python
# 可以看到 如果选择了就餐时付费的话 流失率会随着时间大幅上升
# 另一个就是手工转账的方式 上图稳定的是 花呗付款和微信自动扣款的方式
```

```python
# 现在看看性别对留存率的影响
fig_gender, ax_gender = plt.subplots(figsize=(10,6)) #画布
life_by_cat(feature='性别', ax=ax_gender) #调用函数
ax_gender.set_title('性别对留存的影响') #图题
ax_gender.set_xlabel('入会月数') #X轴标签
ax_gender.set_ylabel('留存率(%)') #Y轴标签
plt.show() #显示图片
```

```python
# 以上也看得出来，无论男生女生，对用户的存留影响并不是很明显
# 到这里就通过按照特征类别分组的留存曲线显示除了哪些特征是对用户留存率是有影响的哪些特征是没有影响的
```

```python
'''
使用Cox危害系数模型 来分析每一个字段对于生存的影响
然后预测某一个人大概多久之后就会流失，并给出随时间延长，生存的概率值大小，就是可能性

'''
```

```python
# Cox危害系数模型可读取的格式是虚拟变量编码 所以还需要进行数据预处理
category_features = ['会员卡类型','会费支付方式']
df_member = pd.get_dummies(df_member,
                          drop_first=True,
                          columns=category_features)
df_member
```

```python
# 创建Cox危害系数模型 并拟合
cph = lifelines.CoxPHFitter()
```

```python
cph.fit(df_member, duration_col='入会月数',event_col='已停付会费', show_progress=False)
```

```python
# 显示留存概率曲线
cph.predict_survival_function(df_member.loc[3]).plot()
```

```python
'''
图表说明：
对这个用户来说，初始的留存概率当然是 1.0；在注册会员 10 个月之后，他留存下来的可能性降低到 50% 左右；
在注册会员 20 个月之后，他留存的概率减少到了 20% 以下；
30 个月之后，这个用户就极有可能会流失，此时的留存概率几乎为 0。
'''
```

```python
# 找到每一个指标（特征字段）对用户的流失影响程度
fig, ax = plt.subplots(figsize=(12,7))
ax.set_title('各个特征的留存相关系数')
cph.plot(ax=ax)
```

```python
# 上图看来大于0的系数值预示着对留存有负面影响，值越大影响越大
# 而小于0的系数值则对留存有正面影响，值越小，影响越大

'''
综上可以看出， 像月卡、手工转账支付会员费这种，都是潜在的流失的原因，这说明“易速鲜花”应该努力让用户办长期卡
用自动付费取代手工转账
'''
```
