### 易速鲜花播放量预测

```python
# 第一步定义问题
"""
目的是估计浏览量，在这个数据集中：点赞数 转发数 热度指数 文章评级 这4个字段都是特征
浏览量就是标签 我们已经有了要估计的标签了 
所以这是一个*监督学习问题*，加上标签是一个连续性的数值，因此它是一个 *回归问题*
"""
```

```python
# 第二步 收集数据和预处理
"""
收集数据
数据可视化
数据清洗
特征工程
构建特征集和标签集
拆分训练集、验证集和测试集
"""
#数据可视化
import pandas as pd
df_ads = pd.read_csv('易速鲜花微信软文.csv')
df_ads.head(15)
```

```python
# 导入可视化数据的库
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文字体为 SimHei（黑体），解决中文乱码
plt.rcParams['font.family'] = 'SimHei'
# 设置负号正常显示
plt.rcParams['axes.unicode_minus'] = False
plt.plot(df_ads['点赞数'], df_ads['浏览量'], 'r.', label='Traning data')
plt.xlabel('点赞数')
plt.ylabel('浏览量')
#显示图例
plt.legend()
plt.show()
```

```python
# 在生成一个箱线图 箱线图是什么
data = pd.concat([df_ads['浏览量'], df_ads['热度指数']], axis=1)
fig = sns.boxplot(x='热度指数', y='浏览量', data=data)
fig.axis(ymin=0, ymax=800000)
```

```python
# 数据清洗 一般有4种情况
# 怎么处理缺失的数据
# 怎么处理重复的数据
# 怎么处理错误的数据
# 怎么处理不可用的数据

#表中含有NaN 的数据行查看
df_ads.isna().sum()
```

```python
# 删除NaN的数据
df_ads = df_ads.dropna()
```

```python
# 在箱线图种有少量的离群点 建议保留 使模型泛化能力高
```

```python
# 特征工程
'''
特征工程是能够使得模型训练效率更高的一些处理，
'''
```

```python
# 构建特征集和标签集

# 特征集 就是那几个除 浏览量之外的自变量
X = df_ads.drop(['浏览量'], axis=1)
# 标签集 就是我们想要预测的数据
y = df_ads.浏览量

X.head()
y.head()
```

```python
# 拆分训练集 验证集 测试集
# 拆分 留作测试的数据一般是20%或者30%
# 使用scikit-learn来拆分
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# 使用线性回归算法创建模型
from sklearn.linear_model import LinearRegression
linereg_model = LinearRegression()

#对于LinearRegression模型，内部参数它自己调 外部参数有两个
#fit_intercept ，默认值为 True，代表是否计算模型的截距。
#normalize，默认值为 False，代表是否对特征 X 在回归之前做规范化。
```

```python
# 训练模型
linereg_model.fit(X_train, y_train)
```

```python
# 预测测试集的Y值
y_pred = linereg_model.predict(X_test)
```

```python
# 对比真值和预测值
df_ads_pred = X_test.copy()
df_ads_pred['浏览量真值'] = y_test
df_ads_pred['浏览量预测值'] = y_pred
df_ads_pred
```

```python
# 查看当前模型长什么样
print('当前模型的4个特征的权重分别是：', linereg_model.coef_)
print('当前模型的截距（偏置）是：',linereg_model.intercept_)
```

```python
# 线性回归预测评分 R方
print('线性回归预测评分：', linereg_model.score(X_test, y_test))
```
